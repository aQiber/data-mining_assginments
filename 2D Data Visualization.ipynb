```python
# @title Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go
import urllib.request
```

```python
# @title Download dataset
urllib.request.urlretrieve('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 'iris.data')
!ls
```

```python
# @title Reading Data
data = pd.read_csv('/content/iris.data', header=None, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width" , "Class"])

data
```

```python
# @title 1. Discarding the fourth dimension
data = data.drop('Petal Width', axis=1)

data
```

```python
# @title 2. Plotting the data points using the first three dimensions (Sepal Length, Sepal Width, Petal Length) in 3d plots
sepal_length = data.iloc[:,0].to_numpy()
sepal_width = data.iloc[:,1].to_numpy()
petal_length = data.iloc[:,2].to_numpy()
class_info = data.iloc[:,3].to_numpy()

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')

ax.scatter(sepal_length[class_info=='Iris-versicolor'], sepal_width[class_info=='Iris-versicolor'], petal_length[class_info=='Iris-versicolor'], c='r', marker='^', s=20)
ax.scatter(sepal_length[class_info=='Iris-setosa'], sepal_width[class_info=='Iris-setosa'], petal_length[class_info=='Iris-setosa'], c='g', marker='s', s=20)
ax.scatter(sepal_length[class_info=='Iris-virginica'], sepal_width[class_info=='Iris-virginica'], petal_length[class_info=='Iris-virginica'], c='b', marker='o', s=20)

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')
ax.set_zlabel('Petal Length', fontweight ='bold')

plt.show
```

```python
# @title  3. Calculating the mean data point for each class and showing them with similar shape with the larger size
meanX1 = np.mean(sepal_length[class_info=='Iris-versicolor'])
meanX2 = np.mean(sepal_length[class_info=='Iris-setosa'])
meanX3 = np.mean(sepal_length[class_info=='Iris-virginica'])

meanY1 = np.mean(sepal_width[class_info=='Iris-versicolor'])
meanY2 = np.mean(sepal_width[class_info=='Iris-setosa'])
meanY3 = np.mean(sepal_width[class_info=='Iris-virginica'])

meanZ1 = np.mean(petal_length[class_info=='Iris-versicolor'])
meanZ2 = np.mean(petal_length[class_info=='Iris-setosa'])
meanZ3 = np.mean(petal_length[class_info=='Iris-virginica'])

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')

ax.scatter(meanX1, meanY1, meanZ1, c='r', marker='^', s=150)
ax.scatter(meanX2, meanY2, meanZ2, c='g', marker='s', s=150)
ax.scatter(meanX3, meanY3, meanZ3, c='b', marker='o', s=150)

ax.set_xlabel('Mean Sepal Length', fontweight ='bold')
ax.set_ylabel('Mean Sepal Width', fontweight ='bold')
ax.set_zlabel('Mean Petal Length', fontweight ='bold')
ax.legend()

plt.show
```

```python
# @title  4. Plotting histogram for each of the dimension (Sepal Length)
ax = plt.hist(sepal_length[class_info=="Iris-versicolor"], color='r')
ax = plt.hist(sepal_length[class_info=="Iris-setosa"], color='g')
ax = plt.hist(sepal_length[class_info=="Iris-virginica"], color='b')
plt.title("Sepal Length")
plt.figure(figsize=(5, 5))

ax.legend()
plt.show
```

```python
# @title  4. Plotting histogram for each of the dimension (Sepal Width)
ax = plt.hist(sepal_width[class_info=="Iris-versicolor"], color='r')
ax = plt.hist(sepal_width[class_info=="Iris-setosa"], color='g')
ax = plt.hist(sepal_width[class_info=="Iris-virginica"], color='b')
plt.title("Sepal Width")

plt.figure(figsize=(5, 5))
plt.show
```

```python
# @title  4. Plotting histogram for each of the dimension (Petal Length)
ax = plt.hist(petal_length[class_info=="Iris-versicolor"], color='r')
ax = plt.hist(petal_length[class_info=="Iris-setosa"], color='g')
ax = plt.hist(petal_length[class_info=="Iris-virginica"], color='b')
plt.title("Petal Length")

plt.figure(figsize=(5, 5))
plt.show()
```

```python
# @title  5. Calculating standard deviation (sigma) for each of the dimensions
sigma_X = np.std(sepal_length)
sigma_Y = np.std(sepal_width)
sigma_Z = np.std(petal_length)

print("Sepal Length: ", sigma_X)
print("Sepal Width: ", sigma_Y)
print("Petal Length: ", sigma_Z)
```

```python
# @title 6. Draw box plot for all data points
ax = plt.figure()
plt.figure(figsize=(5, 5))

labels = ['Sepal Length', 'Sepal Width', 'Petal Length']
data = [sepal_length, sepal_width, petal_length]
ax = plt.boxplot(data, labels = labels)

plt.title("Box Plots for All Data Points")
```

```python
# @title 7. Drawing box plots for each class separately (Iris-setosa)
plt.figure(figsize=(5, 5))
labels = ['Sepal Length', 'Sepal Width', 'Petal Length']
data = [sepal_length[class_info=="Iris-setosa"], sepal_width[class_info=="Iris-setosa"], petal_length[class_info=="Iris-setosa"]]
ax = plt.boxplot(data, labels = labels)

plt.title("Iris-setosa")
```

```python
# @title 7. Drawing box plots for each class separately (Iris-versicolor)
plt.figure(figsize=(5, 5))
labels = ['Sepal Length', 'Sepal Width', 'Petal Length']
data = [sepal_length[class_info=="Iris-versicolor"], sepal_width[class_info=="Iris-versicolor"], petal_length[class_info=="Iris-versicolor"]]
ax = plt.boxplot(data, labels = labels)

plt.title("Iris-versicolor")
```

```python
# @title 7. Drawing box plots for each class separately (Iris-virginica)
plt.figure(figsize=(5, 5))
labels = ['Sepal Length', 'Sepal Width', 'Petal Length']
data = [sepal_length[class_info=="Iris-virginica"], sepal_width[class_info=="Iris-virginica"], petal_length[class_info=="Iris-virginica"]]
ax = plt.boxplot(data, labels = labels)

plt.title("Iris-virginica")
```

```python
# @title 8. Make matrix plots for the entire dataset for 3 dimensions (3x3 scatter matrices)
df = pd.DataFrame({'Sepal Length':sepal_length,
                   'Sepal Width':sepal_width,
                   'Petal Length':petal_length})

pd.plotting.scatter_matrix(df, alpha = 0.2, figsize = (6, 6), marker='+')
```

```python
# @title 8. Plotly giving straight lines only in the diagonals
df = {'Sepal Length':sepal_length,
      'Sepal Width':sepal_width,
      'Petal Length':petal_length}

fig = px.scatter_matrix(df)
fig.show()
```

```python
# @title 9. Drawing parallel coordinates for the entire dataset
fig = go.Figure(data=
    go.Parcoords(
        line_color='blue',
        dimensions = list([
            dict(range = [0,8],
                 label = 'Sepal Length', values = sepal_length),
            dict(range = [0,8],
                 label = 'Sepal Width', values = sepal_width),
            dict(range = [0,8],
                 label = 'Petal Length', values = petal_length)
        ])
    )
)
fig.show()
```

```python
# @title 9. Visually observing the mean and std for each dimension and matching with previous calculation
observed_mean_sepal_length = 5.75
calculated_mean_sepal_length = np.mean(sepal_length)

observed_mean_sepal_width = 3.25
calculated_mean_sepal_width = np.mean(sepal_width)

observed_mean_petal_length = 4
calculated_mean_petal_length = np.mean(petal_length)

print("Observered & Experimental Mean")
print("Sepal Length: ", observed_mean_sepal_length, calculated_mean_sepal_length, f" | Error = {round((abs(calculated_mean_sepal_length - observed_mean_sepal_length)/observed_mean_sepal_length*100),2)}%")
print("Sepal Width: ", observed_mean_sepal_width, calculated_mean_sepal_width, f" | Error = {round((abs(calculated_mean_sepal_width - observed_mean_sepal_width)/observed_mean_sepal_width*100),2)}%")
print("Petal Length: ", observed_mean_petal_length, calculated_mean_petal_length, f" | Error = {round((abs(calculated_mean_petal_length - observed_mean_petal_length)/observed_mean_petal_length*100),2)}%")
print()

print("Observered & Experimental Standard Deviation")
print("Visually observed")
print("Standard Deviation of Petal Length > Sepal Length > Sepal Width")
print("Calculated Values")
print("Sepal Length: ", np.std(sepal_length))
print("Sepal Width: ", np.std(sepal_width))
print("Petal Length: ", np.std(petal_length))
```
