#2D

```python
# @title 1. Plotting the data points using the first two dimensions (Sepal Length and Sepal Width)
sepal_length = data.iloc[:,0].to_numpy()
sepal_width = data.iloc[:,1].to_numpy()
petal_length = data.iloc[:,2].to_numpy()
class_info = data.iloc[:,3].to_numpy()

plt.scatter(sepal_length, sepal_width)
plt.show()
```

```python
# @title 2.  Plotting three different shapes (triangle, square, circle) to plot datapoints for three different classes
ax = plt.axes()

ax.scatter(sepal_length[class_info=='Iris-versicolor'], sepal_width[class_info=='Iris-versicolor'], c='aqua', marker='^', label='Iris-versicolor')
ax.scatter(sepal_length[class_info=='Iris-setosa'], sepal_width[class_info=='Iris-setosa'], c='lightsalmon', marker='s', label='Iris-setosa')
ax.scatter(sepal_length[class_info=='Iris-virginica'], sepal_width[class_info=='Iris-virginica'], c='violet', marker='o', label='Iris-virginica')

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')
ax.legend()

plt.show()
```

```python
# @title 3. Calculating the mean data point for each class and showing them with similar shape with the larger size
ax = plt.axes()

ax.scatter(sepal_length[class_info=='Iris-versicolor'], sepal_width[class_info=='Iris-versicolor'], c='aqua', marker='^', label='Iris-versicolor')
ax.scatter(sepal_length[class_info=='Iris-setosa'], sepal_width[class_info=='Iris-setosa'], c='lightsalmon', marker='s', label='Iris-setosa')
ax.scatter(sepal_length[class_info=='Iris-virginica'], sepal_width[class_info=='Iris-virginica'], c='violet', marker='o', label='Iris-virginica')

meanX1 = np.mean(sepal_length[class_info=='Iris-versicolor'])
meanX2 = np.mean(sepal_length[class_info=='Iris-setosa'])
meanX3 = np.mean(sepal_length[class_info=='Iris-virginica'])

meanY1 = np.mean(sepal_width[class_info=='Iris-versicolor'])
meanY2 = np.mean(sepal_width[class_info=='Iris-setosa'])
meanY3 = np.mean(sepal_width[class_info=='Iris-virginica'])

ax.scatter(meanX1, meanY1, c='blue', marker='^', s=100, label = 'Mean Iris-versicolor')
ax.scatter(meanX2, meanY2, c='orangered', marker='s', s=100, label = 'Mean Iris-setosa')
ax.scatter(meanX3, meanY3, c='purple', marker='o', s=100, label = 'Mean Iris-virginica')

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')
ax.legend(bbox_to_anchor=(1,1.025), loc="upper left", borderpad=2)

plt.show()
```

```python
# @title 4. Plotting using centred vesion of mean
meanLength = np.mean(sepal_length)
meanWidth = np.mean(sepal_width)

centered_sepal_length = sepal_length - meanLength
centered_sepal_width = sepal_width - meanWidth

print(centered_sepal_length)
print(centered_sepal_width)
```

```python
# @title 5. Plotting a line 'l' with line equation (x1 = - x2)
ax = plt.axes()

ax.scatter(centered_sepal_length[class_info=='Iris-versicolor'], centered_sepal_width[class_info=='Iris-versicolor'], c='aqua', marker='^', label='Iris-versicolor')
ax.scatter(centered_sepal_length[class_info=='Iris-setosa'], centered_sepal_width[class_info=='Iris-setosa'], c='lightsalmon', marker='s', label='Iris-setosa')
ax.scatter(centered_sepal_length[class_info=='Iris-virginica'], centered_sepal_width[class_info=='Iris-virginica'], c='violet', marker='o', label='Iris-virginica')

ax.axline((-2.75, 2.75), slope=-1, c='r')

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')

ax.legend(bbox_to_anchor=(1,1.025), loc="upper left", borderpad=2)
ax.set_xlim(-1.7,2.2)
ax.set_ylim(-1.2, 1.5)
plt.grid()
plt.show()
```

```python
# @title 6. Calculating the projection of each data points on the line l And plotting the projected point on the line using the same shape
ax = plt.axes()

l = np.array([-2.75, 2.75])
points = np.array([centered_sepal_length, centered_sepal_width])
points = points.T
p_versi = points[class_info=='Iris-versicolor']
p_setosa = points[class_info=='Iris-setosa']
p_virginica = points[class_info=='Iris-virginica']

projection_versicolor = np.reshape((np.dot(l,p_versi.T)/np.dot(l,l)),(1,50)).T * np.reshape(l,(1,2))
projection_setosa = np.reshape((np.dot(l,p_setosa.T)/np.dot(l,l)),(1,50)).T * np.reshape(l,(1,2))
projection_virginica = np.reshape((np.dot(l,p_virginica.T)/np.dot(l,l)),(1,50)).T * np.reshape(l,(1,2))

ax.axline((-2.75, 2.75), slope=-1, c='r')

ax.scatter(centered_sepal_length[class_info=='Iris-versicolor'], centered_sepal_width[class_info=='Iris-versicolor'], c='aqua', marker='^', label='Iris-versicolor')
ax.scatter(centered_sepal_length[class_info=='Iris-setosa'], centered_sepal_width[class_info=='Iris-setosa'], c='lightsalmon', marker='s', label='Iris-setosa')
ax.scatter(centered_sepal_length[class_info=='Iris-virginica'], centered_sepal_width[class_info=='Iris-virginica'], c='violet', marker='o', label='Iris-virginica')

ax.scatter(projection_versicolor[:,0], projection_versicolor[:,1], c='darkcyan', marker='^', s=20, label='Projected Iris-versicolor')
ax.scatter(projection_setosa[:,0], projection_setosa[:,1], c='chocolate', marker='s', s=20, label='Projected Iris-setosa')
ax.scatter(projection_virginica[:,0], projection_virginica[:,1], c='darkviolet', marker='o', s=20, label='Projected Iris-virginica')

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')

ax.legend(bbox_to_anchor=(1,1.025), loc="upper left", borderpad=2)
ax.set_xlim(-1.7,2.2)
ax.set_ylim(-1.2, 1.5)
plt.grid()
plt.show()
```

```python
# @title 4. Calculating the centered version of (Sepal Length, Sepal Width, Petal Length)
meanSLength = np.mean(sepal_length)
meanSWidth = np.mean(sepal_width)
meanPLength = np.mean(petal_length)

centered_sepal_length = sepal_length - meanSLength
centered_sepal_width = sepal_width - meanSWidth
centered_petal_length = petal_length - meanPLength
```

```python
# @title 5. Plotting the plane (with yellow plane) spanned by two normal vector
fig = plt.figure(figsize=(7,15))
ax = plt.axes(projection='3d')

#Plotting the points

ax.scatter(centered_sepal_length[class_info=='Iris-versicolor'], centered_sepal_width[class_info=='Iris-versicolor'], centered_petal_length[class_info=='Iris-versicolor'], c='aqua', marker='^', label='Iris-versicolor')
ax.scatter(centered_sepal_length[class_info=='Iris-setosa'], centered_sepal_width[class_info=='Iris-setosa'], centered_petal_length[class_info=='Iris-versicolor'], c='lightsalmon', marker='s', label='Iris-setosa')
ax.scatter(centered_sepal_length[class_info=='Iris-virginica'], centered_sepal_width[class_info=='Iris-virginica'], centered_petal_length[class_info=='Iris-versicolor'], c='violet', marker='o', label='Iris-virginica')

#Plotting the plane

v1 = np.array([1, -2, 1])
v2 = np.array([2, 1, 0])
v3 = np.cross(v1, v2) #Finding the vector that is orthogonal to the two vectors given
p = np.array([0,0,0]) #Choosing a point on the plane

xx, yy = np.meshgrid(range(-3, 4), range(-3, 4))
z = (-v3[0] * xx - v3[1] * yy - np.dot(v3, p)) * 1. / v3[2] #Equation of plane spanned by two normal vectors
#v3[0] * (x - p[0]) + v3[1] * (y - p[1]) + v3[2] * (z - p[2]) = 0
#Rearranging this equation of plane to get Z

ax.plot_surface(xx, yy, z, color='yellow', alpha=0.5)

basis_points_1 = np.array([p + t*v1 for t in np.linspace(-1.5, 1.5, 100)])
ax.plot(basis_points_1[:,0], basis_points_1[:,1], basis_points_1[:,2])

basis_points_2 = np.array([p + t*v2 for t in np.linspace(-1.5, 1.5, 100)])
ax.plot(basis_points_2[:,0], basis_points_2[:,1], basis_points_2[:,2])

ax.set_xlabel('Sepal Length', fontweight ='bold')
ax.set_ylabel('Sepal Width', fontweight ='bold')
ax.set_zlabel('Petal Length', fontweight ='bold')

ax.legend(bbox_to_anchor=(1,1.025), loc="upper left", borderpad=2)

ax.view_init(30, 45)

plt.show()
```

```python
# @title 6. Calculating the projection of each data points on the line 'l'
fig = plt.figure(figsize=(7,15))
ax = plt.axes(projection='3d')

v1 = np.array([1, -2, 1])
v2 = np.array([2, 1, 0])
v3 = np.cross(v1, v2)
p = np.array([0,0,0])

xx, yy = np.meshgrid(range(-3, 4), range(-3, 4))
z = (-v3[0] * xx - v3[1] * yy - np.dot(v3, p)) * 1. / v3[2]

ax.plot_surface(xx, yy, z, color='yellow', alpha=0.5)

#Projection of Data onto Plane

point = np.array([centered_sepal_length, centered_sepal_width, centered_petal_length]).T
p_versi = point[class_info=='Iris-versicolor']
p_setosa = point[class_info=='Iris-setosa']
p_virginica = point[class_info=='Iris-virginica']

pvc3D1 = np.reshape((np.dot(p_versi, v1.T)/np.dot(v1, v1)),(1,50)).T * np.reshape(v1,(1,3))
pvc3D2 = np.reshape((np.dot(p_versi, v2.T)/np.dot(v2, v2)),(1,50)).T * np.reshape(v2,(1,3))

pst3D1 = np.reshape((np.dot(p_setosa, v1.T)/np.dot(v1, v1)),(1,50)).T * np.reshape(v1,(1,3))
pst3D2 = np.reshape((np.dot(p_setosa, v2.T)/np.dot(v2, v2)),(1,50)).T * np.reshape(v2,(1,3))

pvr3D1 = np.reshape((np.dot(p_virginica, v1.T)/np.dot(v1, v1)),(1,50)).T * np.reshape(v1,(1,3))
pvr3D2 = np.reshape((np.dot(p_virginica, v2.T)/np.dot(v2, v2)),(1,50)).T * np.reshape(v2,(1,3))

projection_versicolor3D = pvc3D1 + pvc3D2
projection_setosa3D = pst3D1 + pst3D2
projection_virginica3D = pvr3D1 + pvr3D2

ax.scatter(projection_versicolor3D[:,0], projection_versicolor3D[:,1], projection_versicolor3D[:,2], c='darkcyan', marker='^', s=20, label='Projected Iris-versicolor')
ax.scatter(projection_setosa3D[:,0], projection_setosa3D[:,1], projection_setosa3D[:,2], c='chocolate', marker='s', s=20, label='Projected Iris-setosa')
ax.scatter(projection_virginica3D[:,0], projection_virginica3D[:,1], projection_virginica3D[:,2], c='darkviolet', marker='o', s=20, label='Projected Iris-virginica')

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')

ax.view_init(30, 0)

ax.legend(bbox_to_anchor=(1,1.025), loc="upper left", borderpad=2)
plt.grid()
plt.show()
```

```python
# @title 1. Computing the multivariate mean vector
attributes = data.iloc[:,0:11].to_numpy()

m_variate_mean = np.mean(attributes[:,0:10], axis=0)
print(m_variate_mean)
print("Labels: " + str(attributes[:,10]))
```

```python
# @title 2. Computing the sample covariance matrix as inner products between the columns of the centered data matrix
c_variate_mean = attributes[:,0:10] - m_variate_mean
c_variate_mean = c_variate_mean.astype(float)
print(c_variate_mean.shape)
cov_matrix = np.zeros((10,10))
cov_matrix = np.dot(c_variate_mean.T, c_variate_mean)/c_variate_mean.shape[0]
print(cov_matrix)
```

```python
# @title 3. Compute the sample covariance matrix as outer product between the centered data points
cov_matrix2 = np.zeros((10,10))

for i in range(len(c_variate_mean)):
    cov_matrix2 += np.outer(c_variate_mean[i], c_variate_mean[i])

cov_matrix2 /= c_variate_mean.shape[0]
print(cov_matrix2)
```

```python
# @title 4. Computing the correlation between Attributes 1 and 2 by computing the cosine of the angle between the centered attribute vectors
X1 = c_variate_mean[1]
X2 = c_variate_mean[2]
cosine_angle_X12 = cosine_angle_X12 = np.dot(X1.T, X2) / (np.linalg.norm(X1) * np.linalg.norm(X2))

print(cosine_angle_X12)

ax = plt.axes()
plt.scatter(attributes[:,1], attributes[:,2], s=5, marker='^')
ax.set_xlabel('Attribute 1', fontweight ='bold')
ax.set_ylabel('Attribute 2', fontweight ='bold')
plt.show()
```
